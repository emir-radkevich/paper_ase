```python
import numpy as np
import pandas as pd
from matplotlib import rc
import os
from adjustText import adjust_text
from scipy.stats import kendalltau, pearsonr, spearmanr

import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 10)
```

```python
# Тут перечислены все гены, которые не являются супер дуплицированными
df_nsd = df.query("gene not in @df_segm_dupl.gene")
df_nsd = df_nsd.reset_index(drop=True)
```

```python
# Потенциально импринтированные гены
df_nsd_imp = df_nsd.query("gene in @df_imprinted.Gene")
df_nsd_imp = df_nsd_imp.reset_index(drop=True)
df_nsd_imp
```

### Посмотрим на корреляцию между частотой асэ в норме и опухоли.

```python
def kendall_pval(x,y):
    return kendalltau(x,y)[1]

def pearsonr_pval(x,y):
    return pearsonr(x,y)[1]

def spearmanr_pval(x,y):
    return spearmanr(x,y)[1]

# print("Corr coef is " + str(df_trial["ASE_l"].corr(df_trial["ASE_t"], method="spearman")))
# print("p-val is " + str(df_trial["ASE_l"].corr(df_trial["ASE_t"], method=spearmanr_pval)))
print("Corr coef is " + str(df_nsd["med_MAF_l"].corr(df_nsd["med_MAF_t"], method="spearman")))
print("p-val is " + str(df_nsd["med_MAF_l"].corr(df_nsd["med_MAF_t"], method=spearmanr_pval)))
```

### Сделаем картинку, где будут отображены гены, где частота АСЭ в опухоли выше, чем в норме

```python
def df_bar(tis_type):
    
    if tis_type == "t": tis_type = "ASE_t/ASE_l"; sort_v = "ASE_t"
    if tis_type == "l": tis_type = "ASE_l/ASE_t"; sort_v = "ASE_l"
    df = df_nsd.query('{} > 2 & med_MAF_l >=0.7 & med_MAF_t >=0.7'.format(tis_type))
    df_s = df.sort_values('{}'.format(sort_v), ascending=False)
    df_s = df_s.reset_index(drop=True)

    s_l,nat_col=[],[]
    for i in df_s.iterrows():
        nat_col.append(i[1][0])
        sub = i[1][0] + " (" + str(i[1][3]) + "/" + str(i[1][9]) + ")"
        s_l.append(sub)

    df_s.replace(nat_col, s_l, inplace=True)
    return df_s
```

```python
df_tum = df_bar("t")
```

```python
sns.set(style="whitegrid")

fs=14

plt.rc('text', usetex=True)
plt.rc('font', **{'family': 'serif', "size":fs})
plt.rc('text.latex', unicode=True)        
plt.rcParams['text.latex.preamble'] = [r'\usepackage[russian,english]{babel}', \
                                       r'\usepackage[T2A]{fontenc}']

f, ax = plt.subplots(figsize=(5, 10))

sns.set_color_codes("muted")
sns.barplot(x='ASE_f_t', y="gene", data=df_tum,
            label="число случаев АСЭ\nв опухоли", color='r')

sns.set_color_codes("muted")
sns.barplot(x="ASE_f_l", y="gene", data = df_tum,
            label="число случаев АСЭ\nв печени", color="pink")

plt.ylabel("Ген и число (неопухолевая/опухолевая ткани) образцов с АСЭ", fontsize=fs)
plt.xlabel("Доля образцов с АСЭ, \%", fontsize=fs)
sns.despine()

ax.legend(bbox_to_anchor=(1.5, .5), fontsize=fs-1)

plt.savefig(os.path.join(pic_path, 'high_tumor_ase_frequency_new.png'), dpi=300, bbox_inches='tight')
plt.show()
```
