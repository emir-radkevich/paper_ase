```python
import allel
import pandas as pd
import os
import subprocess
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import plotly.graph_objects as go
```

```python
"""
Рисуем различные диаграммы Венна
"""



def mutect2_s1_s2(patient):
        
#     fig=plt.figure()
    path_to_s1 = f"/media/emir/Storage/Cancer/mutect/output/pat{patient}/s1"
    path_to_s2 = f"/media/emir/Storage/Cancer/mutect/output/pat{patient}/s2"
    mutect2_name = "output_transcr_filtered.vcf"
    s1=allel.vcf_to_dataframe(os.path.join(path_to_s1, mutect2_name),
                              fields=["CHROM", "POS"],
                              alt_number=1)
    s2=allel.vcf_to_dataframe(os.path.join(path_to_s2, mutect2_name),
                              fields=["CHROM", "POS"],
                              alt_number=1)
    merged = pd.merge(s1, s2, left_on=["CHROM","POS"], 
                      right_on=["CHROM","POS"], how="inner")
    venn2(subsets=(s1.shape[0],s2.shape[0],merged.shape[0]),
          set_labels=(f"Mut2_pat{patient}s1",f"Mut2_pat{patient}s2"))
    plt.savefig(f"/media/emir/Storage/Cancer/mutect/output/pat{patient}/pics/mut2_s1_s2.png",
                dpi=250)
    plt.clf()
#     return fig

    

def rvboost_s1_s2(patient):
    
#     fig=plt.figure()
    path_to_s1 = f"/media/emir/Storage/Cancer/mutect/rvboost/{patient}s1"
    path_to_s2 = f"/media/emir/Storage/Cancer/mutect/rvboost/{patient}s2"
    rvboost_s1_name = f"intersect.{patient}_s1_0.0.vcf"
    rvboost_s2_name = f"intersect.{patient}_s2_0.0.vcf"
    s1=allel.vcf_to_dataframe(os.path.join(path_to_s1, rvboost_s1_name),
                              fields=["CHROM", "POS"],
                              alt_number=1)
    s2=allel.vcf_to_dataframe(os.path.join(path_to_s2, rvboost_s2_name),
                              fields=["CHROM", "POS"],
                              alt_number=1)
    merged = pd.merge(s1, s2, left_on=["CHROM","POS"], 
                      right_on=["CHROM","POS"], how="inner")
    venn2(subsets=(s1.shape[0],s2.shape[0],merged.shape[0]),
          set_labels=(f"RVB_pat{patient}s1",f"RVB_pat{patient}s2"))
    plt.savefig(f"/media/emir/Storage/Cancer/mutect/output/pat{patient}/pics/rvb_s1_s2.png",
                dpi=250)
    plt.clf()
#     return fig
    
    
    
def rvboost_mutect2(patient,sample):
    
#     fig=plt.figure()
    path_to_mutect2 = f"/media/emir/Storage/Cancer/mutect/output/pat{patient}/s{sample}"
    mutect2_name = f"output_transcr_filtered.vcf"
    mutect2=allel.vcf_to_dataframe(os.path.join(path_to_mutect2, mutect2_name),
                                   fields=["CHROM", "POS", "FILTER_PASS"],
                                   alt_number=1)
    path_to_rvboost = f"/media/emir/Storage/Cancer/mutect/rvboost/{patient}s{sample}"
    rvboost_name = f"intersect.{patient}_s{sample}_0.0.vcf"
    rvboost = allel.vcf_to_dataframe(os.path.join(path_to_rvboost, rvboost_name), 
                                     fields=["CHROM", "POS"], 
                                     alt_number=1)
#     mutect2.drop(mutect2[mutect2["FILTER_PASS"]==False].index, inplace=True)
#     mutect2.shape
    merged = pd.merge(mutect2, rvboost, left_on=["CHROM","POS"], 
                      right_on=["CHROM","POS"], how="inner")
    venn2(subsets=(mutect2.shape[0],rvboost.shape[0],merged.shape[0]),
          set_labels=(f"Mut2_pat{patient}s{sample}",f"RVB_pat{patient}s{sample}"))
    plt.savefig(f"/media/emir/Storage/Cancer/mutect/output/pat{patient}/pics/rvb_mut2_s{sample}.png",
                dpi=250)
    plt.clf()
#     return fig



"""
Сохраняем все картинки
"""
    

    
def save_pics(patient):
    
    mutect2_s1_s2(patient)
    rvboost_s1_s2(patient)
    [rvboost_mutect2(patient,sample) for sample in range(1,3)]
    
    
    
"""
Рисуем таблицу с количеством транскриптов в mutect2 и rvboost
"""    
    
    
    
def calculate_table(patient, sample):
    
    path_to_files = f"/media/emir/Storage/Cancer/mutect/rvboost/{patient}s{sample}"
    file_name = f"intersect.{patient}_s{sample}_0.0.vcf"
    callset=allel.vcf_to_dataframe(os.path.join(path_to_files, file_name),
                                   fields="*", alt_number=1)
    gene_list=callset["SNPEFF_TRANSCRIPT_ID"].unique().tolist()
    for i in gene_list:
        if isinstance(i,float):
            gene_list.remove(i)
    
    result=subprocess.run("DIR_OUT=/media/emir/Storage/Cancer/mutect; \
                wc -l < $DIR_OUT/rvboost/{0}s{1}/genes_pat{0}s{1}_refGene.bed".format(patient, sample), 
                shell = True, stdout=subprocess.PIPE)
    
    return [len(gene_list), int(result.stdout.decode('utf-8').strip())]



def draw_table(patient):
    path_to_tables=f"/media/emir/Storage/Cancer/mutect/output/pat{patient}"
    result=[calculate_table(patient,sample) for sample in range(1,3)]
    df=pd.DataFrame(result, columns=["RVBOOST", "Mutect2"],
                   index=["s1", "s2"])
    df.to_csv(os.path.join(path_to_tables, f"pat{patient}.csv"))
    return df
    
    
    
"""
Показываем картинки
"""    
    
    
    
def show_pics(patient):
    
    path_to_pics=f"/media/emir/Storage/Cancer/mutect/output/pat{patient}/pics"
    titles=["mut2", "rvb", "rvb vs mut2 (s1)", "rvb vs mut2 (s2)"]
    names=["mut2_s1_s2.png", "rvb_s1_s2.png", "rvb_mut2_s1.png", "rvb_mut2_s2.png"]
    fig=plt.figure(figsize=(18,12))
    columns = 2
    rows = 2
    for i in range(1, columns*rows+1):
        img = plt.imread(os.path.join(path_to_pics, names[i-1]))
        ax=fig.add_subplot(rows, columns, i)
        ax.set_title(titles[i-1])
        ax.set_xticks([]); ax.set_yticks([])
        plt.imshow(img)
```

```python
draw_table("T")
```
